```python
import os
import pandas as pd
from collections import Counter
import math
```

```python
df1 = pd.read_csv("Desc.csv")
df2 = pd.read_csv("Annot.csv")
```

```python
output1 = pd.merge(df1, df2, 
                   on='Name', 
                   how='inner')
```

```python
output1
```

```python
def counter_cosine_similarity(c1, c2):
    terms = set(c1).union(c2)
    dotprod = sum(c1.get(k, 0) * c2.get(k, 0) for k in terms)
    magA = math.sqrt(sum(c1.get(k, 0)**2 for k in terms))
    magB = math.sqrt(sum(c2.get(k, 0)**2 for k in terms))
    return dotprod / (magA * magB)*100

```

```python
for index, x in output1.iterrows():
    try:
        counterA = Counter(x["Topic_x"])
        counterB = Counter(x["Topic_y"])
        print(counter_cosine_similarity(counterA,counterB))
    except:
        pass
```
